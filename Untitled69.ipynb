```python
iris=load_iris()
```

```python
dir(iris)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target']=iris.target
```

```python
iris.target_names
```

```python
df['flower_names']=df.target.apply(lambda x:iris.target_names[x])
df.head()
```
